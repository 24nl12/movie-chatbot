```python
def test_fun(user_input):
    tokenizer = RegexpTokenizer(r'\w+')
    tokens = tokenizer.tokenize(user_input)
    neg_tok_count, pos_tok_count = 0, 0
    for token in tokens:
        current = sentiment.get(token, 'none')
        if current == 'pos': pos_tok_count += 1
        elif current == 'neg': neg_tok_count += 1                                               
    if pos_tok_count > neg_tok_count: return 1
    elif neg_tok_count > pos_tok_count: return -1
    return 0
```

```python
test_fun(temp)
```

```python
from sklearn.feature_extraction.text import CountVectorizer 
from sklearn import linear_model
```

```python
data = {}
texts, y = util.load_rotten_tomatoes_dataset()
y_labels = [1 if x == 'Fresh' else -1 for x in y]

#use the training data to create the vocab 
count_vectorizer = CountVectorizer(min_df=20, #only look at words that occur in at least 20 docs
                            stop_words='english', # remove english stop words
                            ) 
data['X'] = count_vectorizer.fit_transform(texts).toarray()
data['Y'] = np.array(y_labels)

logistic_regression_classifier = sklearn.linear_model.LogisticRegression(penalty=None)

model = logistic_regression_classifier.fit(data['X'], data['Y'])
acc = float(model.score(data['X'], data['Y']))
print(acc)
```
